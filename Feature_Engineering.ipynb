11.Why do we have to perform EDA before fitting a model to the data?

-->Exploratory Data Analysis (EDA) is essential before fitting a model because it helps us understand the underlying structure, patterns, and anomalies in the data. Through EDA, we can detect issues like missing values, outliers, incorrect data types, or imbalanced classes, which could negatively impact model performance. It also allows us to explore relationships between variables, assess the distribution of features, and check assumptions required by certain models (like linearity or normality). By gaining insights into the data, we can make informed decisions about feature selection, preprocessing, and choosing the right modeling approach, ultimately leading to more accurate and reliable results.
